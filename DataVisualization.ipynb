```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Ã–rnek veri
kategoriler = ['A', 'B', 'C', 'D', 'E']
veri = [20, 35, 10, 30, 15]
```

```python
#Bar chart oluÅŸturma --> yatay ve dikey bar chartlar vardÄ±r. Kategorik - sayÄ±sal iliÅŸkisi
plt.bar(kategoriler, veri) #x kategori iÃ§in, y veri iÃ§in

#Grafik baÅŸlÄ±ÄŸÄ±nÄ± ve eksen etiketlerini ayarlama
plt.title('Ã–rnek Bar Chart')
plt.xlabel('Kategoriler')
plt.ylabel('DeÄŸerler')

#GrafiÄŸi gÃ¶sterme
plt.show()
```

```python
#Historgram --> SayÄ±sal verilerin daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶rselleÅŸtirir.

veri = np.random.randn(1000) #rastgele 1000 adet sayÄ±, standart normal daÄŸÄ±lÄ±ma gÃ¶re (ortalama = 0, standart sapma = 1)
```

```python
#Histogram oluÅŸturma
plt.hist(veri, bins = 20, edgecolor='black') #bins, 20 eÅŸit aralÄ±ÄŸa bÃ¶lÃ¼yor, edgecolor ise sÃ¼tunun kenar Ã§izgisinin rengini deÄŸiÅŸtiriyor

#Grafik baÅŸlÄ±ÄŸÄ±nÄ± ve eksen etiketlerini ayarlama
plt.title('Ã–rnek Histogram')
plt.xlabel('DeÄŸerler')
plt.ylabel('Frekans')

#GrafiÄŸi gÃ¶sterme
plt.show()
```

```python
#Pie Chart --> Genelde kategorik verileri gÃ¶rselleÅŸtirmek iÃ§in kullanÄ±lÄ±r, toplam pasta dilimi %100'dÃ¼r
```

```python
#Ã–rnek veri
category = ['A', 'B', 'C', 'D', 'E']
data = [30, 15, 20, 10, 25]
```

```python
#Pasta grafik oluÅŸturma
plt.pie(data, labels=category, autopct='%1.1lf%%') #labels=category, her dilime etiketler atar. autopct, her dilimin Ã¼zerinde yÃ¼zde deÄŸerini gÃ¶sterir

#Grafik merkezini ayarlama
plt.gca().set_aspect('equal') #Ã‡ember ÅŸeklinde gÃ¶rÃ¼nmesi iÃ§in. gca(), mevcut ekseni alÄ±r. set_aspect('equal'), grafiÄŸin tam bir Ã§ember ÅŸeklinde gÃ¶rÃ¼nmesini saÄŸlar

#Grafik baÅŸlÄ±ÄŸÄ±nÄ± ayarlama
plt.title('Ã–rnek Pasta Grafik')

#Grafik gÃ¶sterme
plt.show()
```

```python
#Scatter Plot --> Ä°ki sayÄ±sal deÄŸiÅŸken arasÄ±ndaki iliÅŸkiyi gÃ¶rselleÅŸtirir.
```

```python
#Ã–rnek veri
x = np.random.rand(100) # 0 dahil 1 hariÃ§ arasÄ±nda sayÄ±lar Ã¼retir.
y = np.random.rand(100)
```

```python
#Scatter plot oluÅŸturma
plt.scatter(x, y,)

#Grafik BaÅŸlÄ±ÄŸÄ±nÄ± ve eksen etiketlerini ayarlama
plt.title('Ã–rnek Scatter Plot')
plt.xlabel('X DeÄŸerleri')
plt.ylabel ('Y DeÄŸerleri')

#Grafik gÃ¶sterme
plt.show()
```

```python
#Heat Map --> Ä°ki kategorik deÄŸiÅŸken arasÄ±ndaki iliÅŸkiyi, korelasyonlarÄ±, aykÄ±rÄ± deÄŸerleri tespit etmeyi saÄŸlar
```

```python
#Ã–rnek veri seti
veri = np.random.rand(10,10) #10x10 boyutunda rastgele bir matris
```

```python
#Heat map oluÅŸturma
sns.heatmap(veri, cmap="YlGnBu") #cmap, Ä±sÄ± haritasÄ±nda kullanÄ±lacak renk plaetini belirler (Yellow - Green - Blue) 

#Grafik baÅŸlÄ±ÄŸÄ±nÄ± ayarlama
plt.title("Ã–rnek Heat Map")

#Grafik gÃ¶sterme
plt.show
```

```python
sns.heatmap(veri, cmap="YlGnBu", annot = True, fmt=".2f", linewidths=0.5, linecolor='gray') #annot, hÃ¼crelere deÄŸerleri yazdÄ±rÄ±r. fmt, deÄŸerlerin formatÄ±nÄ± belirler(deÄŸerlerin ondalÄ±k sayÄ± olarak ve virgÃ¼lden sonra 2 basamak gÃ¶sterilmesini saÄŸlar). linewidths, kutucuklarÄ±n kalÄ±nlÄ±klarÄ±nÄ± belirler. linecolor, kutucuklarÄ±n kenarlarÄ±nÄ±n rengini belirler
```
