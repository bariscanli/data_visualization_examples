```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Örnek veri
kategoriler = ['A', 'B', 'C', 'D', 'E']
veri = [20, 35, 10, 30, 15]
```

```python
#Bar chart oluşturma --> yatay ve dikey bar chartlar vardır. Kategorik - sayısal ilişkisi
plt.bar(kategoriler, veri) #x kategori için, y veri için

#Grafik başlığını ve eksen etiketlerini ayarlama
plt.title('Örnek Bar Chart')
plt.xlabel('Kategoriler')
plt.ylabel('Değerler')

#Grafiği gösterme
plt.show()
```

```python
#Historgram --> Sayısal verilerin dağılımını görselleştirir.

veri = np.random.randn(1000) #rastgele 1000 adet sayı, standart normal dağılıma göre (ortalama = 0, standart sapma = 1)
```

```python
#Histogram oluşturma
plt.hist(veri, bins = 20, edgecolor='black') #bins, 20 eşit aralığa bölüyor, edgecolor ise sütunun kenar çizgisinin rengini değiştiriyor

#Grafik başlığını ve eksen etiketlerini ayarlama
plt.title('Örnek Histogram')
plt.xlabel('Değerler')
plt.ylabel('Frekans')

#Grafiği gösterme
plt.show()
```

```python
#Pie Chart --> Genelde kategorik verileri görselleştirmek için kullanılır, toplam pasta dilimi %100'dür
```

```python
#Örnek veri
category = ['A', 'B', 'C', 'D', 'E']
data = [30, 15, 20, 10, 25]
```

```python
#Pasta grafik oluşturma
plt.pie(data, labels=category, autopct='%1.1lf%%') #labels=category, her dilime etiketler atar. autopct, her dilimin üzerinde yüzde değerini gösterir

#Grafik merkezini ayarlama
plt.gca().set_aspect('equal') #Çember şeklinde görünmesi için. gca(), mevcut ekseni alır. set_aspect('equal'), grafiğin tam bir çember şeklinde görünmesini sağlar

#Grafik başlığını ayarlama
plt.title('Örnek Pasta Grafik')

#Grafik gösterme
plt.show()
```

```python
#Scatter Plot --> İki sayısal değişken arasındaki ilişkiyi görselleştirir.
```

```python
#Örnek veri
x = np.random.rand(100) # 0 dahil 1 hariç arasında sayılar üretir.
y = np.random.rand(100)
```

```python
#Scatter plot oluşturma
plt.scatter(x, y,)

#Grafik Başlığını ve eksen etiketlerini ayarlama
plt.title('Örnek Scatter Plot')
plt.xlabel('X Değerleri')
plt.ylabel ('Y Değerleri')

#Grafik gösterme
plt.show()
```

```python
#Heat Map --> İki kategorik değişken arasındaki ilişkiyi, korelasyonları, aykırı değerleri tespit etmeyi sağlar
```

```python
#Örnek veri seti
veri = np.random.rand(10,10) #10x10 boyutunda rastgele bir matris
```

```python
#Heat map oluşturma
sns.heatmap(veri, cmap="YlGnBu") #cmap, ısı haritasında kullanılacak renk plaetini belirler (Yellow - Green - Blue) 

#Grafik başlığını ayarlama
plt.title("Örnek Heat Map")

#Grafik gösterme
plt.show
```

```python
sns.heatmap(veri, cmap="YlGnBu", annot = True, fmt=".2f", linewidths=0.5, linecolor='gray') #annot, hücrelere değerleri yazdırır. fmt, değerlerin formatını belirler(değerlerin ondalık sayı olarak ve virgülden sonra 2 basamak gösterilmesini sağlar). linewidths, kutucukların kalınlıklarını belirler. linecolor, kutucukların kenarlarının rengini belirler
```
